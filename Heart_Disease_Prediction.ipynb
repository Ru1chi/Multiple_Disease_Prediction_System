The model used isLogistic Regression

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/heart_disease_data.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
#checking for the null values
df.isnull().sum()
```

```python
df.describe()
```

```python
#seperating the dependent and independent variables
x=df.drop(columns='target',axis=1)
y=df['target']
```

```python
x.shape
```

```python
y.shape
```

```python
#now train_test_split the dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
print(x_train.shape,y_train.shape)
```

```python
#now import the model
model=LogisticRegression()
```

```python
#now train the model using the fit method
model.fit(x_train,y_train)
```

```python
#Model evaluation
prediction_on_train_data=model.predict(x_train)
```

```python
training_data_accuracy=accuracy_score(y_train,prediction_on_train_data)
```

```python
print("The accuracy for the training data is:",training_data_accuracy)
```

```python
#for the test dataset
predictin_on_test_data=model.predict(x_test)
```

```python
accuracy_on_test_data=accuracy_score(predictin_on_test_data,y_test)
```

```python
print("The accuracy on the test data is:",accuracy_on_test_data)
```

```python
#building a predictive system
input_data=[58,0,2,120,340,0,1,172,0,0,2,0,2]
```

```python
#chage input as array
input_array=np.asarray(input_data)
```

```python
#now reshape the array
input_reshaped=input_array.reshape(1,-1)
```

```python
#make prediction method
prediction=model.predict(input_reshaped)
print(prediction)

if(prediction[0]==0):
  print("The person does not have disease.")
else:
  print("The person has a disease.")

```
